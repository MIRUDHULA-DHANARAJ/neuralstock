```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.sort_values(['product_id', 'date']).reset_index(drop=True)
```

```python
df['units_sold'] = df.groupby('product_id')['units_sold'].ffill()
```

```python
df['units_sold'].isnull().sum()
```

```python
df[df['stock_on_hand']<0]



```

```python
group_means = df.groupby('product_id')['stock_on_hand'].transform('mean')

```

```python
df['stock_on_hand']=np.where(df['stock_on_hand']<0, group_means, df['stock_on_hand'])
```

```python
df_encoded = pd.get_dummies(df['product_category'],drop_first=False)
```

```python
df.drop('product_category', axis=1, inplace=True)
```

```python
df = pd.concat([df, df_encoded], axis=1)
```

```python
target_cols=["Beauty", "Electronics", "Home", "Sports"]
for col in target_cols:
    df[col] = df[col].astype(int)
```

```python
Q1 = df['unit_price'].quantile(0.25)
Q3 = df['unit_price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df['unit_price'] = df['unit_price'].clip(lower=lower_bound, upper=upper_bound)
```

```python
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
```

```python
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
```
